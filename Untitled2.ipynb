(1) Data Description

This project uses two datasets collected from a  Minecraft research server.
The first dataset, players.csv, contains 196 observations and 7 variables describing each unique player’s  behavior.
The second dataset, sessions.csv, contains 1,535 observations and 5 variables describing individual play sessions.
Two datasets are linked by the key hashedEmail, which identifies each player.

In players.csv, variables include:

experience (categorical): self-reported skill level

subscribe (logical): whether a player subscribed to a newsletter

hashedEmail (string): encrypted unique player ID

played_hours (numeric): total gameplay hours

name (string): player’s name

gender (categorical): reported gender

Age (numeric): player age,, with 2 missing values


In sessions.csv, variables include:

hashedEmail (string): player ID linking sessions to player data.

start_time and end_time (string): timestamp for each session, with 2 missing end times.

original_start_time and original_end_time (numeric): numeric time codes 

Potential issues include minor missing values, right-skewed playtime (a few heavy users).

Data might be automatically collected via server logs, ensuring accuracy but possibly including interrupted sessions.
Overall, the data are clean, consistent, and suitable for regression analysis after merging by hashedEmail and encoding categorical variables as factors.

(2) Questions
Broad Question:
How do player characteristics affect engagement levels in the Minecraft research server?

Specific Question:
Can player age predict total playtime (played_hours) in the dataset?
Specifically, do older or younger players spend significantly different amounts of time on the server when controlling for gender?

The dataset “players.csv” provides each player’s age, gender, and total gameplay hours.
By using linear regression, we can examine how Age will affect played_hours while accounting for gender differences.

We will use only the “players.csv” file.
Missing values in Age (two observations) will be removed.
The variable gender will be converted into a categorical factor, and played_hours will be kept as a numeric variable.
The final tidy dataset will include three columns: Age, gender, and played_hours.

(3) Exploratory Data Analysis and Visualization

This section explores the relationship between player age and total playtime.
In this analysis, we use only players.csv.

```r
library(tidyverse)

players <- read_csv("players.csv")

head(players)
glimpse(players)
```

```r
players_clean <- players %>%
  filter(!is.na(Age)) %>%
  mutate(
    gender = as.factor(gender) )

summary(players_clean)
```

```r
players_clean %>%
  summarise(across(where(is.numeric), function(x) mean(x, na.rm = TRUE)))
```

```r
ggplot(players_clean, aes(x = Age, y = played_hours)) +
  geom_point(alpha = 0.7, color = "steelblue") +
  labs(
    title = "Relationship Between Player Age and Total Play Hours",
    x = "Age (years)",
    y = "Total Play Hours",
    caption = "Source: Minecraft Research Server data"
  ) +
  theme_minimal(base_size = 13)
```

The scatter plot shows that younger players tend to spend more total hours on the Minecraft research server, and older players generally play less. A few younger participants (under 25) show extremely high total playtime above 200 hours. The overall trend appears slightly negative, indicating that total play hours decrease as age increases.

(4) Methods and Plan

To answer the question, I plan to use a multiple linear regression model. This method is appropriate because the response variable (played_hours) is continuous, and we aim to measure how one numeric predictor (Age) and one categorical control variable (gender) explain its variation.

Why this method is appropriate:
Linear regression can show the direction and magnitude of the relationship between age and playtime, when we  controlling for gender differences. It also allows easy interpretation of coefficients (e.g., how many more or fewer hours are associated with each additional year of age).

Assumptions:
The model assumes (1) a linear relationship between predictors and response, (2) normally distributed data

Limitations:
The sample size (n=196) is too small to develop a solid conclusion of the relationship. Outliers (e.g., very active players) could influence regression estimates，and lead to biases.

Comparison
Two models will be compared: one using only Age and another adding gender.
Their predictive performance will be evaluated using Root Mean Squared Prediction Error (RMSPE) on the testing data.
The model with the lower RMSPE will be selected as the better-fitting model.

Data processing plan:
The cleaned dataset (players_clean) will be used. No extra tidy need to be made. Data will be split into 70% training and 30% testing sets to evaluate predictive accuracy. Cross-validation may be considered later during the group project stage.
